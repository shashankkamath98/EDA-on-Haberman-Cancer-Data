```python
#adding the white grid style 
sns.set_style("whitegrid");
#adding datapoint colour and size
sns.FacetGrid(df, hue="Survival_Status",size=5)\
        .map(plt.scatter, "Age_Of_Patient","Axil_nodes")\
        .add_legend(); #legend fuction will describe class color
plt.show();
    
```

```python
#pairplot on Habeman dataset
plt.close()
sns.pairplot(df,hue="Survival_Status",vars=['Age_Of_Patient', 'Operation_Year', 'Axil_nodes'],size=4)
plt.show()
```

```python
def dens_plot(features,class_var):
    #adding the white grid style 
    sns.set_style(style="whitegrid")
    #adding datapoint colour and size
    sns.FacetGrid(data=df, hue=class_var,size=5)\
        .map(sns.distplot,features)\
        .add_legend()
```

```python
dens_plot("Age_Of_Patient","Survival_Status")
plt.show()
```

```python
#Box Plot for the feature axillary nodes
sns.boxplot(x = 'Survival_Status', y ='Axil_nodes', data = df)
plt.show()
#Box Plot for the feature Age
sns.boxplot(x = 'Survival_Status',y = 'Age_Of_Patient', data = df)
plt.show()
#Box Plot for the feature Year of Operations
sns.boxplot(x = 'Survival_Status', y = 'Operation_Year', data = df)
plt.show()
```

```python
#Violin Plots for Axillary_Nodes
sns.violinplot(x = "Survival_Status", y = "Axil_nodes", data = df, size = 8)
plt.show()

#Violin Plots for Operation_Year
sns.violinplot(x = "Survival_Status", y = "Operation_Year", data = df, size = 8)
plt.show()

#Violin Plots for Age
sns.violinplot(x = "Survival_Status", y = "Age_Of_Patient", data = df, size = 8)
plt.show()
```

```python
sns.jointplot(data=df, x="Age_Of_Patient",y="Operation_Year", kind="kde")
plt.ylabel("Cancer")
plt.show()
```
